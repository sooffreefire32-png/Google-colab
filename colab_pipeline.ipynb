```
# CELL 3: Fetch Data
import requests, json, base64
h = {'Authorization': f'token {GITHUB_TOKEN}', 'Accept': 'application/vnd.github.v3+json'}
r = requests.get(f'https://api.github.com/repos/{GITHUB_REPO}/contents/pipeline_data.json', headers=h).json()
data = json.loads(base64.b64decode(r['content']).decode())
print(f"Pipeline Loaded: {len(data['models'])} models ready.")
```

```
# CELL 4: Model Generation
import bpy, os
for m in data['models']:
    bpy.ops.wm.read_factory_settings()
    bpy.ops.object.select_all(action='SELECT')
    bpy.ops.object.delete()
    bpy.ops.mesh.primitive_monkey_add() if m['type']=='character' else bpy.ops.mesh.primitive_cube_add()
    path = f'/content/{m["name"]}.glb'
    bpy.ops.export_scene.gltf(filepath=path)
    with open(path, 'rb') as f: c = base64.b64encode(f.read()).decode()
    requests.put(f'https://api.github.com/repos/{GITHUB_REPO}/contents/{m["name"]}.glb', headers=h, json={'message': 'add model', 'content': c})
    print(f'Pushed {m["name"]}')
```
