```
# Helper Script to generate 3D objects
with open('generate_scene.py', 'w') as f:
    f.write("""import bpy\nimport os\nbpy.ops.wm.read_factory_settings()\nbpy.ops.object.select_all(action='SELECT')\nbpy.ops.object.delete()\nbpy.ops.mesh.primitive_monkey_add(location=(0,0,0))\nbpy.ops.export_scene.gltf(filepath='/content/scene.glb')\n""")

print('Generating 3D Assets...')
!blender -b -P generate_scene.py
print('Rendering Video...')
!ffmpeg -loop 1 -i /content/scene.glb -t 5 -c:v libx264 -vf "scale=1280:720" -pix_fmt yuv420p /content/final_video.mp4
print('Video Done: /content/final_video.mp4')
```
